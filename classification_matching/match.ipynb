```python
import pandas
google_api = pandas.read_excel('/Users/weihongtan/Desktop/google api/GOOGLE API BUSSINESS TYPE NAMES.xlsx',header=0,sheet_name = 'Only_matched')
google_api['GOOGLE API ALL BUSINESS TYPE'] = google_api['GOOGLE API ALL BUSINESS TYPE'].str.strip()
google_api['GOOGLE API ALL BUSINESS TYPE'] = google_api['GOOGLE API ALL BUSINESS TYPE'].str.lower()

```

```python


armadale = pandas.read_excel('armadale_cleaned.xlsx',header=0)
armadale = pandas.DataFrame(armadale)
business_type = armadale['temp_business_types'].str.split(',', expand=False)


classification  = dict(zip(google_api['GOOGLE API ALL BUSINESS TYPE'], google_api['Classification']))
Category  = dict(zip(google_api['GOOGLE API ALL BUSINESS TYPE'], google_api['Category']))
Category_code  = dict(zip(google_api['GOOGLE API ALL BUSINESS TYPE'], google_api['Category code']))
Sub_category  = dict(zip(google_api['GOOGLE API ALL BUSINESS TYPE'], google_api['Sub-category']))
Sub_category_code  = dict(zip(google_api['GOOGLE API ALL BUSINESS TYPE'], google_api['Sub- category code']))
#Description_and_examples  = dict(zip(google_api['GOOGLE API ALL BUSINESS TYPE'], google_api['Description and examples'])
```

```python
business_type = armadale['temp_business_types'].copy()

for i in range(len(business_type)):



    matched = []
    for type in business_type[i].split(','):

        
        if type in classification.keys():
            matched.append(type)
            

            
    business_type[i] = matched
 #   armadale['temp_business_types'][i] =armadale['temp_business_types'][i][0]      
#business_type = business_type.str.strip()
#business_type = business_type.str.lower()
#armadale['temp_business_types']
#business_type = armadale['temp_business_types']
business_type
```

```python
##noot
for i in range(len(armadale)):
    try:
        armadale['classification_name'][i]= classification[business_type[i]]
    except:
        armadale['classification_name'][i]= ''

    try:
        armadale['category_1'][i]= Category[business_type[i]]
    except:
        armadale['category_1'][i]= ''
    
    try:
        armadale['sub_category_1'][i]= Category_code[business_type[i]]
    except:
        armadale['sub_category_1'][i]= ''
    try:
        armadale['category_2'][i]= Sub_category[business_type[i]]
    except:
        armadale['sub_category_2'][i]= ''
    try:
        armadale['category_3'][i]= Sub_category_code[business_type[i]]
    except:
        armadale['category_3'][i]= ''

    
    
```
